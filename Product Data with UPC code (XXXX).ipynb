```python
import pandas as pd
import numpy as np
import re
```

```python
ProdStartdf = pd.read_excel("Downloads\\Product Data with UPC_start_Test.xlsx")
```

```python
ProdStartdf['Column_Z'] = ProdStartdf['Upc Code(House)']
ProdStartdf['Column_Z'] = ProdStartdf['Column_Z'].str[:-1]
```

```python
ProdStartdf['Column_AA'] = ProdStartdf['Column_Z']
ProdStartdf['Column_AA'] = ProdStartdf['Column_AA'].str.zfill(13)
```

```python
ProdStartdf['Column_AB'] = ProdStartdf['Column_Z']
ProdStartdf['Column_AB'] = ProdStartdf['Column_AB'].str.lstrip("0")
```

```python
ProdStartdf['Column_AC'] = ProdStartdf['On Hand (Cases)'] + ProdStartdf['On Order (Cases)']
```

```python
regEx_pattern1 = re.compile(r'(\d\dX\d\d|\dX\d\d)')
```

```python
regEx_pattern2 = re.compile(r'(\d+X\d+|\d\dX\d)')
```

```python
ProdStartdf['Pattern'] = None
```

```python
index_product = ProdStartdf.columns.get_loc('Product')
index_pattern = ProdStartdf.columns.get_loc('Pattern')
```

```python
ProdStartdf_copy = ProdStartdf
```

```python
for row in range(0, len(ProdStartdf)):
    Pattern_output1 = re.search(regEx_pattern1, ProdStartdf.iat[row, index_product])
    ProdStartdf.iat[row, index_pattern] = Pattern_output1
```

```python
for row in range(0, len(ProdStartdf)):
    Pattern_output2 = re.search(regEx_pattern2, ProdStartdf.iat[row, index_product])
    ProdStartdf.iat[row, index_pattern] = Pattern_output2
```

```python
ProdStartdf.Pattern = ProdStartdf.Pattern.astype('str')
```

```python
ProdStartdf[['Pattern_Match1','Pattern_Match2']] = ProdStartdf['Pattern'].str.split("'",n=1,expand=True)
```

```python
ProdStartdf[['Actual_Match','Match3']] = ProdStartdf['Pattern_Match2'].str.split("'",n=1,expand=True)
```

```python
ProdStartdf[['First_Multiple','Second_Multiple']] = ProdStartdf['Actual_Match'].str.split("X",n=1,expand=True)
```

```python
ProdStartdf.First_Multiple = ProdStartdf.First_Multiple.astype('float')
```

```python
ProdStartdf.Second_Multiple = ProdStartdf.Second_Multiple.astype('float')
```

```python
ProdStartdf['Adjusted_U/C'] = ProdStartdf['First_Multiple'] * ProdStartdf['Second_Multiple']
```

```python
print(ProdStartdf.columns.values)
```

```python
ProdGoaldf = ProdStartdf[['Beverage Type', 'Product Type', 'Product Type #', 'Brand', 'Product',
 'Product #', 'Monitor Code', 'Size Descriptor', 'Qty Per Case',
 'Upc Code(House)', 'State Portfolio Manager', 'State Sales Finance Manager',
 'Supplier', 'Supplier #', 'Colorado Sales Division', 'Marketing Agent',
 'Pallet Size', 'Layer Size', 'Weight', 'Metrics', 'On Hand (Cases)',
 'On Order (Cases)', 'Allotted Inventory (Bottles)',
 'Allotted Inventory (Cases)', 'Available Inventory (Cases)',
 'Column_Z', 'Column_AA', 'Column_AB', 'Column_AC', 'Adjusted_U/C', 'Proj 30 Days Sales']]
```

```python
ProdGoaldf['Adjusted_U/C'] = ProdGoaldf['Adjusted_U/C'].fillna(0)
```

```python
ProdGoaldf = ProdGoaldf.replace({
    'Adjusted_U/C' : 0.0
}, 1.0)
```

```python
ProdGoaldf.loc[ProdGoaldf['Adjusted_U/C'] == 1, 'Adjusted_U/C'] = ProdGoaldf['Qty Per Case']
```

```python
ProdGoaldf.rename({'Column_Z': 'UPC (no check digit)', 'Column_AA': 'King Soopers UPC'}, axis=1, inplace=True)
```

```python
ProdGoaldf.rename({'Column_AB': 'UPC (no check digit, as number)', 'Column_AC': 'OHOO (Cases)'}, axis=1, inplace=True)
```

```python
ProdGoaldf.to_excel("Product Data with UPC goal.xlsx",index=False)
```
